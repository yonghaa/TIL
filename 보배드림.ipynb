```python
import pandas as pd
import numpy as np
import re
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver=webdriver.Chrome()
url="https://www.bobaedream.co.kr/cyber/CyberCar.php?gubun=K"
driver.get(url)
```

```python
html=driver.page_source
html
```

```python
soup=BeautifulSoup(html,'html.parser')
soup
```

```python
#차명 주행거리 가격
```

```python
cars=soup.select("div.list-inner>div.mode-cell.title")[0]
soup.select("div.list-inner>div.mode-cell.km")[0].text.strip()
soup.select("div.list-inner>div.mode-cell.price")[0].text.strip().replace("\n", "")
```

```python
cars=soup.select("div.list-inner>div.mode-cell.title")
cars
```

```python
len(soup.select("div.list-inner>div.mode-cell.title"))
```

```python
cars_list=[]
driver=webdriver.Chrome()
for j in range(1,4):
    url = "https://www.bobaedream.co.kr/cyber/CyberCar.php?gubun=K&page=" + str(j) + "&order=S11&view_size=70"
    driver.get(url)
    html = driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    for i in range(len(cars)):
        name = soup.select("div.list-inner>div.mode-cell.title>p")[i].text.strip()
        distance = soup.select("div.list-inner>div.mode-cell.km")[i].text.strip()
        price = soup.select("div.list-inner>div.mode-cell.price")[i].text.strip().replace("\n", "")
        cars_list.append([name,distance,price])

        print(name, distance, price)
```

```python
df=pd.DataFrame(cars_list, columns=["자동차종류","주행거리","가격"])
df
```

```python
df.to_excel('cars_list.xlsx',index=False)
```
