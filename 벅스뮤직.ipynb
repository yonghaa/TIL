```python
import pandas as pd
import numpy as np
import re
```

```python
from selenium import webdriver
```

```python
driver=webdriver.Chrome()
```

```python
url="https://music.bugs.co.kr/chart"
driver.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
html=driver.page_source
soup=BeautifulSoup(html, 'html.parser')
soup
```

```python
len(soup.select("tr"))
```

```python
songs = soup.select("tr")[1:-3]
songs
```

```python
len(songs)
```

```python
songs
```

```python
song1 = songs[0]
song1
```

```python
title = song1.select("p.title > a")[0].text
title
```

```python
singer=song1.select("p.artist > a")[0].text
singer
```

```python
song_data=[]
rank=1
for song1 in songs:
    title=song1.select("p.title > a")[0].text
    singer=song1.select("p.artist > a")[0].text
    song_data.append(['Bugs', rank, title, singer])
    rank+=1
    print(title, singer, sep=" | ")
```

```python
df=pd.DataFrame(song_data, columns=['서비스','순위','제목','가수명'])
df
```
