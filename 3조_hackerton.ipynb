```python
data_month=data[[ 'electricalAppliances', 'month1', 'month2',
       'month3', 'month4', 'month5', 'month6', 'month7', 'month8', 'month9',
       'month10', 'month11', 'month12']]
data_month
```

```python
import pandas as pd

for i in range(1, 32):
    data_month[f'appliance_{i}'] = 0

# Function to safely convert to set of integers, handling empty and non-numeric values
def safe_int_set(s):
    try:
        return set(map(int, s.split(', ')))
    except ValueError:
        return set()

# Function to process each row
def process_row(row):
    appliances = safe_int_set(row['electricalAppliances'])
    for month_col in ['month1', 'month2', 'month3', 'month4', 'month5', 'month6', 
                      'month7', 'month8', 'month9', 'month10', 'month11', 'month12']:
        if pd.notna(row[month_col]):  
            month_usage = safe_int_set(row[month_col])
            for appliance in appliances:
                if appliance in month_usage:
                    row[f'appliance_{appliance}'] += 1
    return row


data_month = data_month.apply(process_row, axis=1)
data_month

```

```python
x=df[:2500]
test=df[2500:]
```

```python
# 데이터 분할하기
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y= train_test_split(x,y, test_size=0.2) 
```

```python
# 모델구축 및 예측하기
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rf= RandomForestRegressor()
param_grid={ 'n_estimators': [50,150,150], 'max_depth': [20,40,60], 'min_samples_split': [2,5,10]}
gs=GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, verbose=2 )
gs.fit(train_x, train_y)
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
pred_y= gs.best_estimator_.predict(val_x)
pred_y
```

```python
# 모델에 대한 평가
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
mse=mean_squared_error(val_y, pred_y)
print("MSE:", mse)
print("RMSE:", np.sqrt(mse))
print("MAE:", mean_absolute_error(val_y, pred_y))
print("R2_SCORE:",  r2_score(val_y, pred_y))
```
