```python
import pandas as pd
import numpy as np
import re
from selenium import webdriver
from bs4 import BeautifulSoup
import konlpy
from konlpy.tag import Okt
from collections import Counter
from kiwipiepy import Kiwi
import seaborn as sns
import matplotlib.pyplot as plt
import wordcloud

#  
import psycopg2

db_connection = psycopg2.connect(
    host="localhost", #ip
    dbname="final_project",
    user="postgres",
    password="sesac0916",
    port=5432
)

cursor = db_connection.cursor()
# SELECT  
cursor.execute("SELECT * FROM final_project.news_information;")
#  
result = cursor.fetchall()
#   (  5 )
#  
cursor.close()
db_connection.close()

#  
df_news = pd.DataFrame(result, columns=['id', 'create_dt', 'media_company', 'news_genre', 'news_image_path', 'news_title', 'news_origin', 'news_url', 'news_summary'])
df_news

# 'create_dt'  datetime  
df_news['create_dt'] = pd.to_datetime(df_news['create_dt'])

# 'news_genre' ''  
df_IT = df_news[df_news['news_genre'] == 'IT/']

# Kiwi   
kiwi = Kiwi()

#     
noun_by_date = df_IT.groupby(df_IT['create_dt'].dt.strftime('%m-%d'))['news_title'] \
    .apply(lambda x: kiwi.tokenize(' '.join(x))) \
    .apply(lambda tokens: [token.form for token in tokens if len(token.form) >= 2 and token.tag in ['NNG', 'NNP', 'NNB', 'NR', 'NP']])

#      
word_counts_by_date = noun_by_date.apply(lambda words: pd.Series(words).value_counts())

# / 
df=word_counts_by_date.T

#       
df_27 = df[['12-27']].sort_values(by='12-27', ascending=False)
df_27 = df_27.reset_index()

#  
df_27.columns = ['word','n']

#  0 
df_27=df_27.fillna(0)

#  
font="C:\Windows\Fonts\HMFMPYUN.TTF"


dic_word=df_27.set_index("word").to_dict()['n']

from wordcloud import WordCloud
wc = WordCloud(random_state = 1234,         #  
               font_path = font,            #  
               width = 400,                 #  
               height = 400,                #  
               background_color = 'white')  # 

img_wordcloud=wc.generate_from_frequencies(dic_word)

#  
plt.figure(figsize = (10, 10))

#   
import PIL
icon = PIL.Image.open('c:/Users/user/Downloads/dragon.png')

img = PIL.Image.new('RGB', icon.size, (255, 255, 255))

img.paste(icon, icon)
img = np.array(img)

wc = WordCloud(random_state = 1234,         #  
               font_path = font,            #  
               width = 400,                 #  
               height = 400,                #  
               background_color = 'white',  # 
               mask = img)                  # mask 

#   
img_wordcloud = wc.generate_from_frequencies(dic_word)

#   
plt.figure(figsize = (10, 10))  # ,   
plt.axis('off')                 #   
plt.imshow(img_wordcloud)       #   
```
