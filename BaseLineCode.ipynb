```python
import re
```

```python
# 전자기기 사용한 시간 정보는 모두 합쳐서 계산
def catch_sum(str):
    try:
        numbers_list = re.findall(r'\d+', str)
        numbers_list = [int(num) for num in numbers_list]
        result = sum(numbers_list)

        return result
        
    except:
        print(str)
        result = None
        return result
```

```python
for i in ['computerUsageTime','tvUsageTime','airconWithHeatUsageTime','airconUsageTime','heaterUsageTime','elecStoveUsageTime','elecBlanketUsageTime']:
    data[i] = data[i].apply(catch_sum)
```

```python
data[['memberNo', 'jobType', 'houseType', 'houseArea', 'location', 'covidEffect', 'covidPeriod']] = data[['memberNo', 'jobType', 'houseType', 'houseArea', 'location', 'covidEffect', 'covidPeriod']].astype(int)
```

## 3. 학습셋과 검증셋 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(labels='power',axis=1)
Y = data['power']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=100)
```

## 4. 알고리즘 선택 및 모델 학습

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

## 5. 모델을 통한 검증 데이터 예측 수행

```python
import math
```

```python
pred = model.predict(X_test)
```

```python
preds = [math.ceil(pred) for pred in pred]
```

```python
model.score(X_test, Y_test)
```
