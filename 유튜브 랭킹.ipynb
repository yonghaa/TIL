```python
from selenium import webdriver
```

```python
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=1"
driver.get(url)
```

```python
html=driver.page_source
soup=BeautifulSoup(html, 'html.parser')
soup
```

```python
len(soup.select("tr"))
```

```python
yout=soup.select("tr")[1:-1]
```

```python
yout1=yout[0]
yout1
```

```python
goodok=yout1.select("td.subscriber_cnt")[0].text
```

```python
View=yout1.select("td.view_cnt")[0].text
```

```python
Video=yout1.select("td.video_cnt")[0].text
```

```python
Cate=yout1.select("p.category")[0].text.strip()
```

```python
Chan=yout1.select("h1>a")[0].text.strip()
```

```python
Yout_data = []
rank = 1
for yout1 in yout:
    Cate=yout1.select("p.category")[0].text.strip()
    Chan=yout1.select("h1>a")[0].text.strip()
    goodok=yout1.select("td.subscriber_cnt")[0].text
    View=yout1.select("td.view_cnt")[0].text
    Video=yout1.select("td.video_cnt")[0].text
    Yout_data.append(['youtoube', rank, Cate, Chan, goodok, View, Video])
    rank += 1
    print(Cate, Chan, goodok, View, Video, sep = " | ")
```

```python
df=pd.DataFrame(Yout_data, columns=['서비스', '순위', '카테고리', '채널명', '구독자수', '조회수', '동영상'])
df
```
