```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
x_train.shape
```

```python
W=torch.zeros(1, requires_grad=True) #requires_grad=True는 W는 업데이트가되어지는 변수다
```

```python
b=torch.zeros(1, requires_grad=True) #requires_grad=True는 b는 업데이트가되어지는 변수다
```

```python
optimizer=optim.SGD([W,b],lr=0.01)
for epoch in range(2000):
    hx=W * x_train + b
    cost=torch.mean((hx - y_train)**2)
    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌,안해주면 이전에 해준기울기가 계속더해짐
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    if epoch%100 ==0:
        print(epoch,  W.item(), b.item(), cost.item())
```

```python
x1_train = torch.FloatTensor([[73], [93], [89], [96], [73]])
x2_train = torch.FloatTensor([[80], [88], [91], [98], [66]])
x3_train = torch.FloatTensor([[75], [93], [90], [100], [70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

```python
x1_train.shape
```

```python
w1 = torch.zeros(1, requires_grad=True)
w2 = torch.zeros(1, requires_grad=True)
w3 = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

```python
optimizer=optim.SGD([w1,w2,w3,b],lr=0.00001)#lr=1e-5랑같음 #0.01 <= lr <= 0.001 보편적인 값이지만 다를수있음 
#1* 10의 -5승 = 1 * 0.00001
for epoch in range(2000):
    hx=w1 * x1_train + w2 * x2_train+ w3 * x3_train + b
    cost=torch.mean((hx - y_train)**2)
    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    if epoch%10 ==0:
        print(epoch, w1.item(), w2.item(), w3.item(), b.item(), cost.item())
```

```python
#                               X1   X2   X3
x_train  =  torch.FloatTensor([[73,  80,  75], 
                               [93,  88,  93], 
                               [89,  91,  80], 
                               [96,  98,  100],   
                               [73,  66,  70]])  
y_train  =  torch.FloatTensor([[152],  [185],  [180],  [196],  [142]])
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
# h(x) = xtrain    w    +  b
# (5x1)= (5x3)   (3x1)  +  b
```

```python
#w1 = torch.zeros(1, requires_grad=True)
#w2 = torch.zeros(1, requires_grad=True)
#w3 = torch.zeros(1, requires_grad=True) #1의 의미 0이 1개다
W = torch.zeros((3,1), requires_grad=True) #3x1행렬 일때 (3,1)로 써준다
b = torch.zeros(1, requires_grad=True)
```

```python
optimizer=optim.SGD([W,b],lr=0.00001)#lr=1e-5랑같음 #0.01 <= lr <= 0.001 보편적인 값이지만 다를수있음 
```

```python
for epoch in range(2000):
    hx=x_train.matmul(W) + b #matmul
    cost=torch.mean((hx - y_train)**2)
    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    if epoch%100 ==0:
        print(epoch, cost.item())
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
model = nn.Linear(1,1) # 입력차원, 출력차원 ex) 1입력되면 2출력 2입력되면 4출력... 하나입력 하나출력
```

```python
list(model.parameters()) #w, b 값
#초기 모델 : hx=0.5153 * x_train -0.4414
```

```python
optimizer=torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(2000):
    hx = model(x_train) #hx=x_train.matmul(W) + b
    cost = F.mse_loss(hx, y_train)   # cost=torch.mean((hx - y_train)**2)
    optimizer.zero_grad()#미분을 해서 구한 기울기를 0으로 초기화해줌
    cost.backward() #W, b에 대한 기울기가 계산
    optimizer.step()#W,b에 대한 업데이트
    
    if epoch%100 ==0:
        print(epoch, cost.item())
```
