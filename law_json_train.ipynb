```python
law = []
for i in range(len(news['documents'])):
  title = news['documents'][i]['title']
  # sentence = news['documents'][i]['text'][0][0]['sentence']
  summary = news['documents'][i]['abstractive'][0]
  a = ''
  for j in range(len(news['documents'][i]['text'][0])):
    sentence = news['documents'][i]['text'][0][j]['sentence']
    a += sentence
  law.append([title,a, summary])
```

```python
df_news = pd.DataFrame(law, columns=['title', 'origin', 'summary'])
df_news
```
