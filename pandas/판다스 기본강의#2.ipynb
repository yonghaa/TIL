```python
#pivot table
#열하나의 데이터를 뽑아서 values로 만든다
#그 values에 index와 columns를 새롭게 만들어서 방향 전환하는 것
#새롭게 만드는 index와 columns도 기존 데이터프레임의 열에서 선택을 한다는 것
```

```python
#1.values / 2.index / 3. columns / 4. aggfunc
```

```python
import pandas as pd
a1 = ['송중기', '이유빈', '권보아', '차유람', '박보검', '김범수',  '이효리', '유승호', '김유정']
a2 = list("남여여여남남여남여")
a3 = list("AAABBBCCC")
a4 = [97, 88, 78, 64, 84, 89, 87, 92, 99]
a5 = [i + "등" for i in list("123321321")]
a6 = [i + "시" for i in list("수정정수수수정정수")]
a7 = ['이름', '성별', '반', '점수', '반등수', '비고']

df = pd.DataFrame([a1, a2, a3, a4, a5, a6], index=a7, columns=list(range(1, 10))).T
df["점수"] = df["점수"].astype(int)
```

```python
df.head(4)
```

```python
#df.pivot_table(열, index=[열1,...,열n], columns=[열1,...,열n], aggfunc=집계함수)
```

```python
df.pivot_table("점수", index="반", columns="성별", aggfunc="mean")
```

```python
df.pivot_table("점수", index=["반","성별"], columns=["비고"], aggfunc="mean")
```

```python
#숫자가 아닌 문자열이 담긴 열을 values로 df를 피벗해보자
#문자열인 경우 집계함수를 쓸 수 없으니까 pivot_table로 안된다. pivot함수 써야한다!!!
```

```python
# pivot함수는 pivot_table과 달리 values도 따로 지정해줘야한다 (Values=열)
```

```python
df.pivot(values=열, index=[열1,...,열n], columns=[열1,...,열n])
```

```python
df.pivot(values="이름", index="반", columns="반등수") #집계함수 지정할필요x
```

```python
#pivot을 하는이유는 데이터를 토대로 한눈에 보기 쉽게 만들기 위해서
#unpivot을 하는이유는 다른방향으로 또다시 pivot을 하기 위해서
#pivot은 다른방향으로 전환시키기가 힘들다 그래서 unpivot했다가 다시 pivot
```

```python
# unpivot
# 언피벗의 순서 1.stack 2.reset_index 3.columns이름을 바꾼다
# 판다스는 stack과 melt로 unpivot을 한다
```

```python
#stack은 index와 columns의 구조를 바꾸는 함수
#stack을 쓰면 columns에서 한줄을 빼서 index로 옮겨준다
#반대로 unstack을하면 index에서 한줄을 빼서 columns로 옮겨준다
```

```python
#stack + reset_index하면 unpivot
```

```python
val = [[83, 93, 99], [77, 88, 91], [92, 78, 95]]
idx = ["권보아", "김범수", "박효신"]
df = pd.DataFrame(val, index=idx, columns=["국어", "수학", "과학"])
val1 = [[20, 10], [5, 25], [22, 8], [12, 18]]
idx1 = [["1반", "1반", "2반", "2반"], ["남", "여","남","여"]]
df1 = pd.DataFrame(val1,index=idx1, columns=["도시락", "샌드위치"])
df2 = df1.copy().unstack()
df3 = df1.copy()
df3["업체"] = ["CJ푸드", "신세계", "신세계", "CJ푸드"]
df4 = df3.reset_index().rename(columns={"level_0":"반", "level_1":"성별"}).copy()
```
